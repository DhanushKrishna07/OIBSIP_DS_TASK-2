```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the iris dataset
iris_sklearn = load_iris()
iris = pd.DataFrame(iris_sklearn.data, columns=iris_sklearn.feature_names)
iris['Species'] = pd.Categorical.from_codes(iris_sklearn.target, iris_sklearn.target_names)
```

```python
# Rename columns to match the original script
iris.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']
```

```python
# Apply the seaborn-whitegrid style
plt.style.use('seaborn-whitegrid')
```

```python
# Set the figure size
plt.figure(figsize=(8, 8))
```

```python
# Calculate the correlation matrix using only numeric columns
correlation_matrix = iris.select_dtypes(include='number').corr()
```

```python
# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, linewidths=.5)

# Show the plot
plt.show()
```
